```python
import ast

def parse_python_literal(value):
    if not value.startswith("[") and not value.startswith("{"):
        raise ValueError("Expected list or dict")
    return ast.literal_eval(value)
```

```python
def calc_duration_seconds(duration_string: str):
    end = False
    seconds = 0
    while not end:
        end = True
        hr_i = duration_string.find("hr")
        if hr_i > -1:
            seconds += int(duration_string[:hr_i - 1]) * 60 * 60
            duration_string = duration_string[hr_i + 2:].strip()
            end = False

        min_i = duration_string.find("min")
        if min_i > -1:
            seconds += int(duration_string[:min_i - 1]) * 60
            duration_string = duration_string[min_i + 3:].strip()
            end = False

        sec_i = duration_string.find("sec")
        if sec_i > -1:
            seconds += int(duration_string[:sec_i - 1])
            duration_string = duration_string[sec_i + 3:].strip()
            end = False
    return seconds
```

```python
df_anime["duration_seconds"] = df_anime["duration_str"].map(calc_duration_seconds)
```

```python
df_anime["total_duration_seconds"] = df_anime.apply(lambda row: row["duration_seconds"] * row["episodes"], axis=1)
df_anime["total_duration_seconds"]
```
