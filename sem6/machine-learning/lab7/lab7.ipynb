```python
import numpy as np
```

```python
def load_data():
    path = r"C:\Programing\sem6\ml\lab7\shalyto.txt"
    
    with open(path, "r", encoding="utf-8") as file:
        lines = list(map(lambda line: line[:-1] if line[-1] == "\n" else line, file.readlines()))
        return lines

raw_data = load_data()
```

```python
def rebuild_notes(data):
    cur_note = ""
    notes = []
    i = 0
    
    def find_next_note_start_index():
        nonlocal i
        note_id = str(len(notes) + 1)
        while i < len(data) and not data[i].startswith(note_id + "."):
            i += 1
        return i
    
    prev_i = find_next_note_start_index()
    
    def next_note():        
        nonlocal i
        nonlocal prev_i
        nonlocal notes
        note_id = str(len(notes) + 1)
        
        if i >= len(data):
            return False
        
        find_next_note_start_index()
        
        def _filter(x):
            return not x.isnumeric()
        
        note = " ".join(filter(_filter, data[prev_i:i]))
        
        prev_i = i
        
        note = note[note.find(".") + 2:]
        notes.append(note)
        return True
    
    while next_note():
        pass
    return notes[1:]
        
notes = rebuild_notes(raw_data)
del raw_data
notes[:50]
```

```python
def _f1(x):
    if "(" in x and not ")" in x:
        return x[:x.find("(")]
    if ")" in x and not "(" in x:
        return x[x.find(")")+1:]
    return x

def split_sentences(notes):
    result = []
    for note in notes:
        _note = note.replace("…", ".").replace("!", ".").replace("?", ".")
        sentenses = map(lambda x: x.strip(), _note.split("."))
        sentenses = map(_f1, sentenses)
        sentenses = map(lambda x: x.strip(), sentenses)
        sentenses = filter(lambda x: len(x) > 10, sentenses)
        result.extend(sentenses)
    return result

sentenses = split_sentences(notes)
del notes
sentenses[:50]
```

```python
sentenses = list(map(lambda x: x.lower(), sentenses))
sentenses[:50]
```

```python
alphabet = set("йцукенгшщзхъфывапролджэячсмитьбюё. ")
len(alphabet)
```

```python
def filter_chars(x):
    result = []
    for ch in x:
        if ch in alphabet:
            result.append(ch)
    return "".join(result) + "."

sentenses = list(map(filter_chars, sentenses))
sentenses[:50]
```

```python
sum(map(len, sentenses))
```

```python
ch_to_i = {ch: i for i, ch in enumerate(sorted(alphabet))}
i_to_ch = {i: ch for i, ch in enumerate(sorted(alphabet))}
ch_to_i, i_to_ch
```

```python
def one_hot(sentenses):
    result = []
    for sentense in sentenses:
        sub_result = []
        for ch in sentense:
            onehot = [0] * len(alphabet)
            onehot[ch_to_i[ch]] = 1
            sub_result.append(onehot)
        result.append(sub_result)
    return result
```

```python
X_raw = one_hot(sentenses)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import LSTM, Dense
```

```python
X = []

for sentense in X_raw:
    X.extend(sentense)
    
len(X)

X = X[:len(X) // 3]

del X_raw
del sentenses
```

```python
from tqdm import tqdm

max_len = 20

X_train, Y_train = [], []
X_test, Y_test = [], []

for i in tqdm(range(len(X) - max_len - 1)):
    if i > len(X) - len(X) // 4:
        X_test.append(X[i:i+max_len])
        Y_test.append(X[i+max_len])
    else:
        X_train.append(X[i:i+max_len])
        Y_train.append(X[i+max_len])

X_train, Y_train = map(np.array, (X_train, Y_train))
X_test, Y_test = map(np.array, (X_test, Y_test))
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)

del X

import gc
gc.collect()
```

```python
lstm_model = keras.Sequential([
    keras.Input(shape=(max_len, len(alphabet))),
    LSTM(128, dropout=0.1),
    Dense(len(alphabet), activation="softmax")
])

lstm_model.summary()
```

```python
lstm_model.compile(loss="categorical_crossentropy", optimizer="adam")
```

```python
batch_size=128
epochs=15

lstm_model.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
# lstm_model.fit(X_train, Y_train, batch_size=batch_size, epochs=5, validation_split=0.1)

loss = lstm_model.evaluate(X_test, Y_test)
print("loss:", loss)
```
