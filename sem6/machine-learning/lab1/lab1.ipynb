# Настройка гиперпараметров

## Расстояния

```python
def minkovsky_dist(p):
    def _dist(x1, x2):
        assert len(x1) == len(x2)
        n = len(x1)
        sum_ = 0.0
        for i in range(n):
            sum_ += abs(x1[i] - x2[i]) ** p
        return sum_ ** (1 / p)
    return _dist
    
euclid_dist = minkovsky_dist(2)
manhattan_dist = minkovsky_dist(1)
```

```python
def chebyshev_dist(x1, x2):
    assert len(x1) == len(x2)
    return max([abs(a - b) for a, b in zip(x1, x2)])
```

```python
distances = [
    ("euclid_dist", euclid_dist),
    ("manhattan_dist", manhattan_dist),
    ("chebyshev_dist", chebyshev_dist)
]
```

## Ядра

```python
kernels = [
    ("uniform_k", lambda u: 0.5 if abs(u) <= 1.0 else 0.0),
    ("triangular_k", lambda u: max(1.0 - abs(u), 0.0)),
    ("epanechnikov_k", lambda u: max(0.75 * (1.0 - u ** 2), 0.0)),
    ("quartic_k", lambda u: max(15 * (1 - u ** 2) ** 2 / 16, 0.0))
]
```

## Типы окна

```python
window_types = [("fixed_w", True), ("non_fixed_w", False)]
```

## Параметры окна

```python
n = len(df)
print("len(df) =", n)
```

```python
def rd(ro):
    max_ = 0.0
    for x1 in X:
        for x2 in X:
            max_ = max(max_, ro(x1, x2))
    return max_
```

```python
def gen(start, stop, step):
    res = []
    x = start
    while x < stop:
        res.append(x)
        x += step
    return res
```

```python
from math import sqrt

theta_by_window_type_by_ro = {
    True: (lambda ro: gen(rd(ro) / sqrt(n), rd(ro), rd(ro) / sqrt(n))),
    False: (lambda _: list(range(1, int(sqrt(n)) + 1)))
}
```

## Leave-one-out

```python
def leave_one_out(X, Y, idx):
    assert(len(X) == len(Y))
    return np.delete(X, idx, axis=0), np.delete(Y, idx, axis=0), X[idx], Y[idx]
```

```python
X_train, Y_train, x_test, y_test = leave_one_out(X, Y, 0)
```

```python
Y_train.shape
```

```python
from regression import nadaraya_watson

def model_fit(X_train, Y_train, ro, K, theta, is_fixed):
    models = []
    for i in range(Y_train.shape[1]):
        models.append(nadaraya_watson(X_train, Y_train[:, i], ro, K, theta, is_fixed))
    return lambda x: np.array([models[i](x) for i in range(Y_train.shape[1])])
```

```python
def count_f1_loo(X, Y, *params):
    tp = 0
    fp = 0
    p = 0
    for i in range(X.shape[0]):
        X_train, Y_train, x_test, y_test = leave_one_out(X, Y, i)
        model = model_fit(X_train, Y_train, ro, ker, param, wtype)
        _y_pred = model(x_test)
        y_pred = np.zeros(len(_y_pred))
        y_pred[_y_pred.argmax()] = 1
        for j in range(len(y_test)):
            if y_test[j] == 1:
                p += 1
                if y_pred[j] == 1:
                    tp += 1
            elif y_pred[j] == 1:
                fp += 1
    
    recall = tp / p
    prec = tp / (tp + fp)
    return 2 * (prec * recall) / (prec + recall)
```

```python
from tqdm import tqdm

best_f1 = 0
best_params = ()
for ro_name, ro in tqdm(distances):
    for ker_name, ker in kernels:
        for wtype_name, wtype in window_types:
            for param in theta_by_window_type_by_ro[wtype](ro):
                print(ro_name, ker_name, wtype_name, param)
                f1 = count_f1_loo(X, Y, ro, ker, param, wtype)
                print(f1)
                if f1 > best_f1:
                    best_f1 = f1
                    best_params = (ro_name, ker_name, wtype_name, param)

print(best_f1, best_params)
```
